### KMeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
feature_set =data.iloc[:,3:12].values
feature_set
```

```python
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters = i,init = 'k-means++',n_init = 10,random_state = 101)
    km.fit(feature_set)
    wcss.append(km.inertia_)
    
plt.figure(figsize = (10,7))                  
plt.plot(range(1,11),wcss)
plt.title("Elbow Method")
plt.xlabel("No of clusters k")
plt.ylabel("WCSS score")
plt.show()
```

Optimum number of clusters is at Elbow point. Here optimum number of clusters, k= 3

```python
km = KMeans(n_clusters = 3,init = 'k-means++',n_init = 10,random_state = 101)
km.fit(feature_set)

data['km_3'] = km.labels_
data
```

```python
plt.figure(figsize = (10,7))                 
plt.scatter(x=data['Magnesium'],y=data['Proanthocyanins'],c=data['km_3'])
plt.show()
```

```python
plt.figure(figsize = (10,7))                 
plt.scatter(x=data['Magnesium'],y=data['Alcohol'],c=data['km_3'])
plt.show()
```

```python
plt.figure(figsize = (10,7))                 
plt.scatter(x=data['Magnesium'],y=data['Total_Phenols'],c=data['km_3'])
plt.show()
```

#### There are three clusters represented by three different colours.

### Hierarchical Clustering

```python
from scipy.cluster.hierarchy import linkage, dendrogram
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
data.loc[:,:] = sc.fit_transform(data)
```

```python
mergings = linkage(data, method = "complete")
```

```python
plt.figure(figsize=(10,6))
dendrogram(mergings) 
plt.title('Dendrogram')
plt.show()
```

```python
mergings.shape
```

```python
from scipy.cluster.hierarchy import fcluster
```

```python
labels = fcluster(mergings,8,criterion = "distance")
```

```python
pd.Series(labels).value_counts()
```

#### From the Dendrogram optimum number of clusters is 5

### Principal Component Analysis

```python
data.corr()
```

```python
from sklearn.decomposition import PCA
pc = PCA(n_components = 8)
```

```python
pc_array = pc.fit_transform(data)
pc_array.shape
```

```python
tranc_data = pd.DataFrame(pc_array,columns = ['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8'])
```

```python
tranc_data.corr().round(2)
```

```python
pc.explained_variance_ratio_
```

#### We can choose 3 as the optimum number of clusters since it contain most of the data.

```python
pc = PCA(n_components = 3)
```

```python
pc_array = pc.fit_transform(data)
pc_array.shape
```

```python
trans_data = pd.DataFrame(pc_array, columns = ["PC1","PC2",'PC3'])
```

```python
trans_data.corr().round(2)
```

```python
pc.explained_variance_ratio_*100
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
mergings = linkage(trans_data, method = "complete")
```

```python
plt.figure(figsize=(10,6))
dendrogram(mergings, labels = data.index)   
plt.show()
```

#### Optimum number of clusters, k = 3
